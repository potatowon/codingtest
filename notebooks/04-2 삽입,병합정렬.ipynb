# 삽입정렬 

````python
array = [4, 6, 2, 9, 1]


def insertion_sort(array):
    for i in range(1, len(array)):
        for j in range(i):
            if array[i-j] < array[i-j-1]:
                array[i-j], array[i-j-1] = array[i-j-1], array[i-j]
            else:
                break # 효율 업
    return array


insertion_sort(array)
print(input) # [1, 2, 4, 6, 9] 가 되어야 합니다!

print("정답 = [4, 5, 7, 7, 8] / 현재 풀이 값 = ",insertion_sort([5,8,4,7,7]))
print("정답 = [-1, 3, 9, 17] / 현재 풀이 값 = ",insertion_sort([3,-1,17,9]))
print("정답 = [-3, 32, 44, 56, 100] / 현재 풀이 값 = ",insertion_sort([100,56,-3,32,44]))
````

# 병합정렬

```
2포인터 이용

1. 왼쪽에서 inx 하나 잡고
2. 오른쪽의 그룹과 비교
3. 왼쪽이 작으면 append 하고 인덱스 더하기
4. 반복 idx = len(array_a) - 1

문제 -> 포인터가 다 돌고 남는게 있으면 넣어주는 절차가 필요.. 


````python
array_a = [1, 2, 3, 5]
array_b = [4, 6, 7, 8]


def merge(array1, array2):
    results = []
    idx = 0
    while idx != len(array1):
        for i in range(len(array2)):
            if array1[idx] > array2[i]:
                results.append(array2[i])
            else:
                results.append(array1[idx])
                break
        idx += 1
    
       # 포인터가 다 돌고나서 붙여주는 값은 어떻게 해결하지?
        # hint 인덱스의 값을 따로 계산 -> 남는게 있다면 따로 넣어주는 과정이 필요하다.....
        # 졸리니까 .. 내일 오전에 다시 생가해보자
    return results


print(merge(array_a, array_b))  # [1, 2, 3, 4, 5, 6, 7, 8] 가 되어야 합니다!

print("정답 = [-7, -1, 5, 6, 9, 10, 11, 40] / 현재 풀이 값 = ", merge([-7, -1, 9, 40], [5, 6, 10, 11]))
print("정답 = [-1, 2, 3, 5, 10, 40, 78, 100] / 현재 풀이 값 = ", merge([-1,2,3,5,40], [10,78,100]))
print("정답 = [-1, -1, 0, 1, 6, 9, 10] / 현재 풀이 값 = ", merge([-1,-1,0], [1, 6, 9, 10]))
````
