- timedelta 객체에는 산술 연산자 +와 -를 사용할 수 있으므로 어떤 날짜에 원하는 기간(일, 시, 분, 초)을 더하거나 뺄 수 있다.

```python
# 243
# datetime.timedelta()는 두 날짜의 차이를 계산할 때 사용하는 함수이다.
# timedelta 객체에는 산술 연산자 +와 -를 사용할 수 있으므로 어떤 날짜에 원하는 기간(일, 시, 분, 초)을 더하거나 뺄 수 있다.


now = datetime.datetime.now()

for day in range(5, 0, -1):
    delta = datetime.timedelta(days=day) # 날짜의 합산을 할 수 있게 만드는 함수
    date = now - delta # + 뒤의 날짜, - 앞의 날짜
    print(date)
```

```python
# 이자 지급하기 
''' 5회다 한 번씩 이자 지급'''


import numpy as np 

class Account:
    # class variable ; 여러 인스턴스 간에 서로 공유해야 하는 값은 클래스 변수를 통해 바인딩해야 합니다
    account_cnt = 0

    def __init__(self, name, balance):
        self.deposit_cnt = 0
        
        self.name = name
        self.balance = balance
        self.back = 'SC 은행'
        
        first = np.random.randint(0, 1000)
        second = np.random.randint(0, 10000)
        third = np.random.randint(0, 1000)
        self.account_number = str(first).zfill(3)+'-'+str(second).zfill(4)+'-'+str(third).zfill(3)
        
        Account.account_cnt += 1
        # print(f'은행이름: SC은행')
        # print(f'계좌번호: {self.account_number}')
    
    @classmethod
    def get_account_num(cls):
        print(cls.account_cnt)

    def deposit(self, money):
       
        if money >= 1:
            self.balance += money

            self.deposit_cnt += 1
        
        if self.deposit_cnt%5 == 0:
            self.balance += self.balance*1.10


    def withdraw(self, account):
        if self.balance >= account:
            self.balance -= account


    def display_info(self):
        print(f'은행이름: {self.back}')
        print(f'예금주: {self.name}')
        print(f'계좌번호: {self.account_number}')
        print(f'잔고: {self.balance:,}원')


p = Account("파이썬", 10000)
p.deposit(10000)
p.deposit(10000)
p.deposit(10000)
p.deposit(5000)
p.deposit(5000)
print(p.balance)
```

```python
## 입출금 내역 확인하기


import numpy as np 

class Account:
    # class variable
    account_cnt = 0

    def __init__(self, name, balance):
        self.deposit_cnt = 0
        self.deposit_log = []
        self.withdraw_log = []

        
        self.name = name
        self.balance = balance
        self.back = 'SC 은행'
        
        first = np.random.randint(0, 1000)
        second = np.random.randint(0, 10000)
        third = np.random.randint(0, 1000)
        self.account_number = str(first).zfill(3)+'-'+str(second).zfill(4)+'-'+str(third).zfill(3)
        
        Account.account_cnt += 1
        # print(f'은행이름: SC은행')
        # print(f'계좌번호: {self.account_number}')
    
    @classmethod
    def get_account_num(cls):
        print(cls.account_cnt)

    def deposit(self, money):
       
        if money >= 1:
            self.balance += money
            self.deposit_log.append(money)

            self.deposit_cnt += 1
        
        if self.deposit_cnt%5 == 0:
            self.balance += self.balance*1.10


    def withdraw(self, account):
        if self.balance >= account:
            self.withdraw_log.append(account)
            self.balance -= account


    def display_info(self):
        print(f'은행이름: {self.back}')
        print(f'예금주: {self.name}')
        print(f'계좌번호: {self.account_number}')
        print(f'잔고: {self.balance:,}원')
    
    
    def withdraw_history(self):
        for amount in self.withdraw_log:
            print(f'출금: {amount}')

    def deposit_history(self):
        for amount in self.deposit_log:
            print(f'입금: {amount}')


k = Account("Kim", 1000)
k.deposit(100)
k.deposit(200)
k.deposit(300)
k.deposit_history()

k.withdraw(100)
k.withdraw(200)
k.withdraw_history()
    
```

```python
'''
try:
    실행코드
except 예외 as 변수:
    예외처리 코드
'''

data = [1, 2, 3]

for i in range(5):
    try:
        print(data[i])
    except IndexError as e:
        print(e)
```

```python
'''
try:
    실행 코드
except:
    예외가 발생했을 때 수행할 코드
else:
    예외가 발생하지 않았을 때 수행할 코드
finally:
    예외 발생 여부와 상관없이 항상 수행할 코드
'''

per = ["10.31", "", "8.00"]

for i in per:
    try:
        print(float(i))
    except:
        print(0)
    else:
        print('clean data')
    finally:
        print('변환완료')
```
