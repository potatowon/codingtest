# 해쉬
해시(다양한 길이를 가진 데이터를 고정된 길이를 가진 데이터로 변경한 값)화한 key값을 통해 `Key-value`를 저장해둔 자료구조.(파이썬의 dictionary)

 - 연관 배열 추가에 사용
 - 해시 테이블은 해시 함수를 사용하여 아래 배열로 계산 -> 데이터의 검색과 저장이 빠름
     - 색인(index)
     - 버킷(bucket)
     - 슬롯(slot)
     
     > 시간 복잡도 O(n)

### 해시함수
임의의 길이를 갖는 메시지를 입력하여 고정된 길이의 해시값을 출력

 - `put(key, value)` : 저장
 - `get(key)` : key 값에 해당하는 value 가져오기

숫자 1,2,3,4,5 있고, 숫자를 3으로 나눈 나머지 값으로 해싱한다.


|숫자|해시값|
|:--:|:--:|
|1|1|
|2|2|
|3|0|
|4|1|
|5|2|

중복되는 해시값의 존재 > 충돌(collision) : 같은 어레이의 인덱스로 매핑되어 데이터를 덮어 써버리는 결과 발생

```python
# 8개의 값으로 해싱 (0-7)
```

```python
class Dict:
    def __init__(self):
        self.item = [None] * 8 # 8개의 자리 초기화
        
    def put(self, key, value):
        index = hash(key) % len(self.items)
        self.items[index] = value
        
    def get(self, key):
        index = hash(key) % len(self.items)
        return self.items[index]
```

### 체이닝 방식 
해시값이 같은 경우 LinkedList 자료구조를 이용하여 해시값 메모리 위치에 Key-Value 데이터를 하나씩 추가해서 서로 연결해 가며 저장

```python
class LinkedTuple:
    def __init__(self):
        self.items = []
    
    def add(self, key, value):
        self.items.append((key, value)) # 리스트 안에 튜플 형태로 저장
        
    def get(self,key): # 튜플안에서 입력한 키 값이랑 같은 경우만 v 도출
        for k, v in self.items:
            if k == key:
                return v
```

```python
def solution(participant, completion):
    d_p = {} # 참여자 초기화
    d_c = {} # 완주자
    d = {}
    answer = ''

    # participant 딕셔너리 화
    for i in participant:
        if d_p.get(i) is None: # 동명이인이 없는 경우 1
            d_p[i] = 1
        else : # 동명이인이 있는 경우 +1 추가
            d_p[i] += 1
    # print(d_p)


    # completion 딕셔너리 화
    for i in completion:
        if d_c.get(i) is None:
            d_c[i] = 1
        else :
            d_c[i] += 1
    # print(d_c)

    # 같은 key 값의 value 의 차이  > 0 이 아닌 값 출력 한것이 완주하지 못한 사람.

    for k in d_p:
        if d_c.get(k) is None: # 완주자에 목록이 없으면 None -> 계산을 위해 None을 0 으로 변환 후 계산
            d_c[k] = 0
            d[k] = d_p[k] - d_c[k]
        else:
            d[k] = d_p[k] - d_c[k]
    # print(d)


    for key in d:
        if d[key] != 0:
            answer += key
    return answer
```
